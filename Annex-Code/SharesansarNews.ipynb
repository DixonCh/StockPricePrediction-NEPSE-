```python
import pandas as pd
from datetime import datetime
```

```python
news_data = pd.read_csv('news_sharesansar.csv')
```

```python
news_data.head()
```

```python
def changeDate(d):
    return datetime.strptime(d, '%A, %B %d, %Y').strftime('%m/%d/%Y')
```

```python
news_data['DateF'] = news_data['Date'].apply(changeDate)
```
